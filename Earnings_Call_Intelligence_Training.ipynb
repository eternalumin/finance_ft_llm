## Step 6: Upload Model to HuggingFace Hub

```python
# Get your HF token from: https://huggingface.co/settings/tokens
from huggingface_hub import HfApi, create_repo

# Login (will prompt for token)
!huggingface-cli login

# Upload model
api = HfApi()
api.upload_folder(
    folder_path="training/outputs/earnings-intelligence-v1",
    repo_id="eternalumin/earnings-intelligence-3b",
    repo_type="model"
)
print("✅ Model uploaded to: https://huggingface.co/eternalumin/earnings-intelligence-3b")
```

## Step 7: Create HuggingFace Space for Demo

1. Go to: https://huggingface.co/new-space

3. **Hardware:** T4-small
4. **Repository name:** `earnings-intelligence`
5. Copy `demo/gradio_app.py` content to your Space's `app.py`
6. Set environment variable: `MODEL_PATH` = `eternalumin/earnings-intelligence-3b`
